```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()

RAPIDAPI_KEY = os.getenv('RAPIDAPI_KEY')

url = "https://yelp-business-api.p.rapidapi.com/search"

querystring = {"query":"Coffee Shops","location":"Chicago","page":"1"}

headers = {
	"x-rapidapi-key": RAPIDAPI_KEY,
	"x-rapidapi-host": "yelp-business-api.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
```

```python
if response.status_code == 200:
    # parse the JSON data
    json_data = response.json()

    try:
        # Convert JSON to dataframe
        df = pd.json_normalize(json_data)
        print('JSON data uploaded successfully')

    except:
        'Data not loaded into dataframe'

    if 'SearchResults' in df.columns:
        # Expands the SearchResults column
        search_results_df = pd.json_normalize(df['SearchResults'])

        # If SearchResuts column is a list of dictionaries
        search_results_df = pd.concat([pd.json_normalize(record) for record in df['SearchResults']], ignore_index=True)

    else:
        print('Column "SearchResults" is not in the DataFrame')

else:
    print(f"Failed to fetch data: {response.status_code}")

# Display the dataframe
display(search_results_df.head())
```
