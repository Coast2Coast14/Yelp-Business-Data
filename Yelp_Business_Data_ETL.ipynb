```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()

RAPIDAPI_KEY = os.getenv('RAPIDAPI_KEY')

url = "https://yelp-business-api.p.rapidapi.com/search"

querystring = {"query": "Coffee Shop", "location": "Chicago", "page": "1"}

headers = {
    "x-rapidapi-key": RAPIDAPI_KEY,
    "x-rapidapi-host": "yelp-business-api.p.rapidapi.com",
}

response = requests.get(url, headers=headers, params=querystring)

if response.status_code == 200:
    # Parse the JSON data
    json_data = response.json()

    try:
        # Convert JSON to DataFrame
        df = pd.json_normalize(json_data)
        print('Data successfully converted to dataframe')

    except:
        'Data not loaded into dataframe'
    

        # Assume 'SearchResults' is the column with nested JSON
    if 'SearchResults' in df.columns:
        # Expand the 'SearchResults' column
        search_results_df = pd.json_normalize(df['SearchResults'])
        
        # If 'SearchResults' is a list of dictionaries
        search_results_df = pd.concat([pd.json_normalize(record) for record in df['SearchResults']], ignore_index=True)
        

    else:
        print("Column 'SearchResults' not found in the DataFrame")

    # Display the DataFrame
    display(search_results_df.head())
else:
    print(f"Failed to fetch data: {response.status_code}")
```

```python
search_results_df['neighborhoods_2'] = search_results_df['neighborhoods'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
```
